```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces

%matplotlib inline
```

```python
rng = np.random.RandomState(0)
faces, _ = fetch_olivetti_faces(return_X_y=True, shuffle=True, random_state=rng)
```

```python
n_samples, n_features = faces.shape
```

```python
faces.shape
```

```python
#| export
def plot_random_face(faces_dset):
    """Plot a randomly selected face from the Olivetti faces dataset
    
    faces_dset: the scikit-learn dataset containing the faces.
    """
    img = faces_dset[np.random.randint(low=0, high=400, size=1)]
    plt.imshow(img.reshape(64, 64))
    plt.show()
```

```python
plot_random_face(faces)
```

```python
faces.shape
```

```python
def plot_average_face(faces_dset, fn):
    """Plot the average or median face"""
    if fn == 'average':
        avg_face = faces_dset.mean(axis=0)
    elif fn == 'median':
        avg_face = np.median(faces_dset, axis=0)
    plt.imshow(avg_face.reshape(64, 64))
```

```python
plot_average_face(faces, 'average')
```

```python
plot_average_face(faces, 'median')
```
